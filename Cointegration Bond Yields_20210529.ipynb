# Durbin Watson Test

Durbin WatsonTest; used to detect the presence of autocorrelation in the residuals of a regression  
The null hypothesis of the test is that there is no serial correlation

For r == 0, indicating no serial correlation, the test statistic equals 2. This statistic will
always be between 0 and 4. The closer to 0 the statistic, the more evidence for positive serial correlation. The closer to 4, the more evidence for negative serial correlation.

```python
from statsmodels.stats.stattools import durbin_watson
var_list = df_clean.columns

for i in var_list:
    dw = durbin_watson(df_clean[i])    
    dw_diff = durbin_watson(np.diff(df_clean[i]))
    print('Durbin_Watson \t\t ' + i + ':\t' + str("%.3f" %dw))
    print('Durbin_Watson_diff \t ' + i + ':\t' + str("%.3f" %dw_diff)+'\n')
```

# ADF: Augmented Dickey Fuller Test

- Testing the non-differenced data for 1 up to 2 lags
- Testing the differenced data for 1 up to 2 lags


```python
#create df for 1 lag data
df_lag_1 = pd.DataFrame(dtype = 'object',index = ["Stat", "P-Value", "Lags", "Nb of Obs", "CritValues"])

for i in var_list:
    #for lags in range(1,3):
        #print('Number of lags used:',lags)

        adf_library = adfuller(df_clean[i], maxlag=1, regression='nc',autolag=None)
        df_lag_1[i] = adf_library

#Values adf_library[4] for row "CritValues" are dictionaries. Therefore we append rows for each key and its value
crit_values = adf_library[4]

for key in crit_values:
    df_lag_1 = df_lag_1.append(pd.Series(dtype = 'object',name = key))
    df_lag_1.loc[key] = [crit_values[key]]

#we can now drop the old row "CritValues"    
df_lag_1 = df_lag_1.drop("CritValues")  
```

```python
#create df for 2 lags data
df_lag_2 = pd.DataFrame(index = ["Stat", "P-Value", "Lags", "Nb of Obs", "CritValues"])

for i in var_list:
    #for lags in range(1,3):
        #print('Number of lags used:',lags)

        adf_library = adfuller(df_clean[i], maxlag=2, regression='nc',autolag=None)
        df_lag_2[i] = adf_library

#Values adf_library[4] for row "CritValues" are dictionaries. Therefore we append rows for each key and its value
crit_values = adf_library[4]

for key in crit_values:
    df_lag_2 = df_lag_2.append(pd.Series(dtype = 'object', name = key))
    df_lag_2.loc[key] = [crit_values[key]]
    
#we can now drop the old row "CritValues"
df_lag_2 = df_lag_2.drop("CritValues")   
```

```python
#create df for 1 lag difference data
df_lag_1_d = pd.DataFrame(index = ["Stat", "P-Value", "Lags", "Nb of Obs", "CritValues"])


for i in var_list:
    #for lags in range(1,3):
        #print('Number of lags used:',lags)

        adf_library_d = adfuller(np.diff(df_clean[i]), maxlag=1, regression='nc',autolag=None)
        df_lag_1_d[i] = adf_library_d

#Values adf_library[4] for row "CritValues" are dictionaries. Therefore we append rows for each key and its value        
crit_values = adf_library_d[4]

for key in crit_values:
    df_lag_1_d = df_lag_1_d.append(pd.Series(dtype = 'object', name = key))
    df_lag_1_d.loc[key] = [crit_values[key]]

#we can now drop the old row "CritValues"    
df_lag_1_d = df_lag_1_d.drop("CritValues")    
    
```

```python
#create df for 2 lag difference data
df_lag_2_d = pd.DataFrame(index = ["Stat", "P-Value", "Lags", "Nb of Obs", "CritValues"])


for i in var_list:
    #for lags in range(1,3):
        #print('Number of lags used:',lags)

        adf_library_d = adfuller(np.diff(df_clean[i]), maxlag=2, regression='nc',autolag=None)
        df_lag_2_d[i] = adf_library_d

#Values adf_library[4] for row "CritValues" are dictionaries. Therefore we append rows for each key and its value
        
crit_values = adf_library_d[4]

for key in crit_values:
    df_lag_2_d = df_lag_2_d.append(pd.Series(dtype = 'object', name = key))
    df_lag_2_d.loc[key] = [crit_values[key]]

#we can now drop the old row "CritValues"  
df_lag_2_d = df_lag_2_d.drop("CritValues") 
```

# Cointegration Tests using ts.coint

The Null hypothesis is that there is NO cointegration, the alternative hypothesis is that there is a cointegrating relationship. If the pvalue is small, below a critical size, then we can reject the hypothesis that there
is no cointegrating relationship.

The cointegration tests are run on non-differentiated data and will be stored in a dictionary

```python
coin_dict ={}
variables = df_clean.columns
lags = 3

for lag in range(0,lags,1):
    coin_dict[lag] = {}
    for a, b in itertools.permutations(variables, 2):
        x1=df_clean[a]
        y1=df_clean[b]
        coin_result = ts.coint(x1, y1, maxlag = lag) 
        coin_dict[lag][a + ' / ' + b] = coin_result
        
coin_dict
```

Overview for cointegration p-value results

```python

cvariables = df_clean.columns
lags = 3
check = 'abc'
total_counter = 0
coin_count = 0
lagcountlist = []

for lag in range(0, lags, 1):
    print('\n Lags: ' + str(lag) + '\n')
    lag_counter = 0
    for a, b in itertools.permutations(cvariables, 2):
        if not a == check:
            print('\033[1m \n Cointegration results for ' +a + '\n \033[0m')
            
            
        y0=df_clean[a]
        y1=df_clean[b]
        coin_result = ts.coint(y0, y1,  maxlag = lag)
        
        total_counter  = total_counter + 1
        
        
       
    
        if coin_result[1] < 0.05:
            print('cointegrates with \t' + b + '\t (P-Value:' + "%.5f" % coin_result[1] + ')')
            coin_count = coin_count + 1
            lag_counter = lag_counter + 1
            
        check = a
    lagcountlist.append(lag_counter)    

cointlags = str(int(total_counter / 3))    
    
print('\n'+  str(coin_count) + ' out of ' + str(total_counter) + ' cointegration pairs with P-Value below 0.05 \n')    
print(str(lagcountlist[0]) + ' significant cointegration results out of ' + cointlags + ' cointegration tests  at 0 lags')
print(str(lagcountlist[1]) + ' significant cointegration results out of ' + cointlags + ' cointegration tests  at 1 lag')
print(str(lagcountlist[2]) + ' significant cointegration results out of ' + cointlags + ' cointegration tests  at 2 lags')
```

# Granger Causality Tests

The Null hypothesis for grangercausalitytests is that the time series in the second column, x2, does NOT Granger cause the time series in the first column, x1. Granger causality means that past values of x2 have a statistically significant effect on the current value of x1, taking past values of x1 into account as regressors. We reject the null hypothesis that x2 does not Granger cause x1 if the pvalues are below a desired size of the test.

Granger tests will be run on differentiated data with one lag. Results of the tests are saved in a dictionary by default,
out of which we save the p-value from the ssr_ftest into a matrix in a dataframe

```python
gvariables = df_diff.columns
dfgranger = pd.DataFrame()

for a, b in itertools.permutations(gvariables, 2):

    resu = grangercausalitytests(df_diff[[a, b]], maxlag=1, addconst=True, verbose=False)
    p_value = resu[1][0]['ssr_ftest'][1]
    colname = b + '_x'
    indexname = a + '_y'
    dfgranger.at[indexname, colname] = p_value


dfgranger
```
